```python
import numpy as np
import pandas as pd
from googletrans import Translator
```

# Train.csv

```python
data = pd.read_csv("../../project_watson/data/train.csv")
data['lang_abv'].value_counts()
```

```python
data_en = data[data['lang_abv'] == "en"]
data_en.head()
```

```python
data_en_1 = data_en[0:1000]
data_en_2 = data_en[1000:2000]
data_en_3 = data_en[2000:3000]
data_en_4 = data_en[3000:4000]
data_en_5 = data_en[4000:5000]
data_en_6 = data_en[5000:]
```

```python
def translate(pair_of_sentences, dest=None):
    dest_choices = [
        "zh-cn",
        "ar",
        "fr",
        "sw",
        "ur",
        "vi",
        "ru",
        "hi",
        "el",
        "th",
        "es",
        "de",
        "tr",
        "bg",
    ]
    if not dest:
        dest = np.random.choice(dest_choices)

    translator = Translator()
    
    decoded = [translator.translate(words, dest=dest).text for words in pair_of_sentences] + [dest]
    return decoded
```

```python
random_translations = data_en_1.apply(lambda row: translate([row['premise'], row['hypothesis']]), axis=1)
data_en_1['premise'] = [translation[0] for translation in random_translations]
data_en_1['hypothesis'] = [translation[1] for translation in random_translations]
data_en_1['lang_abv'] = [translation[2] for translation in random_translations]
data_en_1.head()
```

```python
random_translations = data_en_2.apply(lambda row: translate([row['premise'], row['hypothesis']]), axis=1)
data_en_2['premise'] = [translation[0] for translation in random_translations]
data_en_2['hypothesis'] = [translation[1] for translation in random_translations]
data_en_2['lang_abv'] = [translation[2] for translation in random_translations]
data_en_2.head()
```

```python
random_translations = data_en_3.apply(lambda row: translate([row['premise'], row['hypothesis']]), axis=1)
data_en_3['premise'] = [translation[0] for translation in random_translations]
data_en_3['hypothesis'] = [translation[1] for translation in random_translations]
data_en_3['lang_abv'] = [translation[2] for translation in random_translations]
data_en_3.head()
```

```python
random_translations = data_en_4.apply(lambda row: translate([row['premise'], row['hypothesis']]), axis=1)
data_en_4['premise'] = [translation[0] for translation in random_translations]
data_en_4['hypothesis'] = [translation[1] for translation in random_translations]
data_en_4['lang_abv'] = [translation[2] for translation in random_translations]
data_en_4.head()
```

```python
random_translations = data_en_5.apply(lambda row: translate([row['premise'], row['hypothesis']]), axis=1)
data_en_5['premise'] = [translation[0] for translation in random_translations]
data_en_5['hypothesis'] = [translation[1] for translation in random_translations]
data_en_5['lang_abv'] = [translation[2] for translation in random_translations]
data_en_5.head()
```

```python
random_translations = data_en_6.apply(lambda row: translate([row['premise'], row['hypothesis']]), axis=1)
data_en_6['premise'] = [translation[0] for translation in random_translations]
data_en_6['hypothesis'] = [translation[1] for translation in random_translations]
data_en_6['lang_abv'] = [translation[2] for translation in random_translations]
data_en_6.head()
```

```python
data_not_en = data[data['lang_abv'] != 'en']
```

```python
train = data_en_1.append([data_en_2, data_en_3, data_en_4, data_en_5, data_en_6, data_not_en], ignore_index=True)
train
```

```python
train['lang_abv'].value_counts()
```

```python
train["language"] = train["lang_abv"].map({
    "ur": "Urdu",
    "ar": "Arabic",
    "zh": "Chinese",
    "zh-cn" : "Chinese",
    "hi": "Hindi",
    "ru": "Russian",
    "de": "German",
    "bg": "Bulgarian",
    "tr": "Turkish",
    "es": "Spanish",
    "en": "English",
    "vi": "Vietnamese",
    "el": "Greek",
    "sw": "Swahili",
    "th": "Thai",
    "fr": "French"
    })
```

```python
train
```

```python
train.to_csv('./train.csv', sep='|', index=False)
```
